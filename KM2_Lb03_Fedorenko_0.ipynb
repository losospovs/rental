### 3.1A
$\color{red}\text{Создайте анимацию}$ построения графика некоторой явно заданной
аналитической функции $y=y(x)$ по значениям $x$, последовательно изменяющимся
от $x_{min}$ до $x_{max}$ с шагом $step$. Выполните задание для $y(x)=sin(x)^4$, $x_{min}=1$, $x_{max}=5$, $step=0.05$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib widget
```

```python
def y(x):
    return np.sin(x)**4
```

```python
x_min=1
x_max=5
step=0.05
```

```python
x_array=np.arange(x_min,x_max,step)
y_array=y(x_array)
```

```python
matrix0=[x_array,y_array]
matrix=np.transpose(matrix0)
```

```python
fig1=plt.figure()
```

```python
y_min, y_max=np.min(y_array),np.max(y_array)
ax1=plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])
```

```python
line1,=ax1.plot([],[],'b',lw=5)
print(type(line1))
line1.get_xdata(),line1.get_ydata(),
```

```python
def at_frame1(point):
    """добавляет к объекту line1 точку с координатами (point[0],point[1])

 Arguments : 
 point : массив из двух элементов

 Returns : None
 """
    x_coord,y_coord=list(line1.get_xdata()),list(line1.get_ydata())
    x_coord.append(point[0])
    y_coord.append(point[1])
    line1.set_data(x_coord,y_coord)
?line1.set_data
```

### 3.1B
$\color{red}\text{Создайте анолгичную анимацию}$ построения графика параметрически
заданной функции $x=x(t)$, $y=y(t)$ по значениям $t$, последовательно
изменяющимся от $t_{min}$ до $t_{max}$ с шагом $step$. Выполните задание для функции из Задания 1.4 в Лб1, согласно Вашему варианту. Значения и выберите самостоятельно с учетом непрерывности параметрической функции на выбранном диапазоне, $step=\frac{t_{max}-t_{min}}{100}$

```python
def x(t):
    return np.log(np.sqrt((1 - t)/(1+t)))
def y(t):
    return np.sqrt(1-t**2)
```

```python
t_min=-0.99
t_max=0.99
step1=(t_max-t_min)/100
t_array1=np.arange(t_min,t_max,step1)
```

```python
x_array1=x(t_array1)
y_array1=y(t_array1)
matrix1=np.transpose([x_array1,y_array1])
```

```python
fig0=plt.figure()
x_min1,x_max1=np.min(x_array1),np.max(x_array1)
y_min1,y_max1=np.min(y_array1),np.max(y_array1)
ax0=plt.axes()
plt.axis([x_min1-0.5, x_max1+0.5, y_min1-0.5, y_max1+0.5])
```

```python
line0,=ax0.plot([],[],'g')
line0.get_ydata()
```

```python
def at_frame0(point):
 """добавляет к объекту line3 точку с координатами (point[0],point[1])

 Arguments :
 point : массив из двух элементов

 Returns : None
 """
 x_coord = list(line0.get_xdata())
 y_coord = list(line0.get_ydata())

 x_coord.append(point[0])
 y_coord.append(point[1])

 line0.set_data(x_coord, y_coord)
```

```python
fig0=plt.figure()
ax1=plt.axes()
plt.axis([x_min1-0.5, x_max1+0.5, y_min1-0.5, y_max1+0.5])
line0,=ax0.plot([],[],'g')
FuncAnimation(fig0, at_frame0, frames=matrix1,repeat=False,interval=15)
```

```python
plt.close('all')
```

## Задание 3.2. Движение точки по линии на плоскости

$\color{red}\text{Создайте анимацию}$ движения точки по графику функции $x=x(t)$, $y=y(t)$ по значениям $t$, последовательно
изменяющимся от $t_{min}$ до $t_{max}$ с шагом $step$. Функция $x=x(t)$, $y=y(t)$ и значения переменных $t_{min}$, $t_{max}$, $step$ задаются такими же, как и при выполнении Задания 3.1b

```python
fig2 = plt.figure()
ax2 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])
line1, = ax2.plot(x_array, y_array,'b') # объект для графика функции
line2, = ax2.plot([], [],'go') # объект для точки
```

```python
def at_frame2(point):
 """задает объект line2 точкой с координатами (point[0],point[1])

 Arguments :

 point : массив из двух элементов

 Returns : None
 """
 line2.set_data([point[0]],[point[1]])
```

```python
fig2 = plt.figure()
ax2 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])

def init():
    # создает начальное состояние графической области
    line1, = ax2.plot(x_array, y_array,'b') # объект для графика функции
    line2, = ax2.plot([], [],'go') # объект для точки
    plt.legend([r'$y(x) = \sin(x)^4$','Movable point'])
FuncAnimation(fig2,at_frame2,frames=matrix,init_func=init,repeat=False,interval=20)
```
